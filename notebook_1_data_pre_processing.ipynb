```python
import pandas as pd
import numpy as np
```

```python
df_2017 = pd.read_excel('Fires_100.xlsx')
```

```python
df_2016 = pd.read_excel('Fires_100.xlsx',1)
```

```python
df_2017.head(5)
```

```python
df_2017.rename(columns={'YEAR_':'YEAR'}, inplace=True)
df_2017.YEAR.fillna(2017, inplace=True)
```

```python
df_2016 = df_2016[df_2016.YEAR!='<Null>'].dropna(subset='YEAR')
```

```python
df_2016.tail(5)
```

```python
df_2016.info()
```

```python
df_2017.info()
```

```python
df = pd.concat([df_2016, df_2017])
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df[['YEAR', 'AGENCY', 'UNIT_ID', 'FIRE_NAME', 'INC_NUM',
       'ALARM_DATE', 'CONT_DATE', 'CAUSE', 'REPORT_AC',
       'GIS_ACRES', 'C_METHOD', 'OBJECTIVE']]
```

```python
df = df.replace('<Null>', np.nan)
df.to_csv('wildfires.csv', index=False)
```
